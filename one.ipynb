```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_billboard():

    driver = webdriver.Chrome()

    driver.get("https://www.billboard.com/charts/hot-100/")

    # Wait for the page to load
    time.sleep(5)

    # Song Titles
    song_elements = driver.find_elements(
        By.CSS_SELECTOR,
        "h3#title-of-a-story"
    )

    # Artists
    artist_elements = driver.find_elements(
        By.CSS_SELECTOR,
        "a[href*='/artist/']"
    )

    # Ranks
    rank_elements = driver.find_elements(
        By.CSS_SELECTOR,
        "span.c-label"
    )

    print("Songs:", len(song_elements))
    print("Artists:", len(artist_elements))
    print("Ranks:", len(rank_elements))

    n = min(len(song_elements), len(artist_elements), len(rank_elements))

    songs = []

    for i in range(n):

        songs.append({

            "rank": rank_elements[i].text.strip(),

            "song": song_elements[i].text.strip(),

            "artist": artist_elements[i].text.strip()

        })

    driver.quit()

    return songs
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_billboard():

    driver = webdriver.Chrome()

    driver.get("https://www.billboard.com/charts/hot-100/")

    time.sleep(5)

    cards = driver.find_elements(
        By.CSS_SELECTOR,
        "li.o-chart-results-list__item"
    )

    billboard = []

    rank = 1

    for card in cards:

        try:
            song = card.find_element(
                By.CSS_SELECTOR,
                "h3#title-of-a-story"
            ).text.strip()

            artist = card.find_element(
                By.CSS_SELECTOR,
                "a[href*='/artist/']"
            ).text.strip()

            billboard.append({

                "rank": rank,
                "song": song,
                "artist": artist

            })

            rank += 1

        except:
            continue

    driver.quit()

    return billboard
```

```python
# startup.py

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time


URL = "https://techcrunch.com/category/startups/"


def fetch_startups():

  
    driver = webdriver.Chrome()

    driver.get(URL)

    # wait for JS
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, "html.parser")

    driver.quit()

    startups = []

    articles = soup.find_all("article")

    for article in articles:

        # -------------------------
        # Title
        # -------------------------
        title_tag = article.find("a", class_="loop-card__title-link")

        if not title_tag:
            continue

        title = title_tag.get_text(strip=True)
        article_url = title_tag.get("href")

        # -------------------------
        # Author
        # -------------------------
        author_tag = article.find("a", class_="loop-card__author")

        if author_tag:
            author = author_tag.get_text(strip=True)
            author_url = author_tag.get("href")
        else:
            author = "N/A"
            author_url = ""

        # -------------------------
        # Category
        # -------------------------
        category_tag = article.find("a", class_="loop-card__category")

        if category_tag:
            category = category_tag.get_text(strip=True)
        else:
            category = "Startup"

        # -------------------------
        # Time
        # -------------------------
        time_tag = article.find("time")

        if time_tag:
            published = time_tag.get_text(strip=True)
        else:
            published = ""

        startups.append({
            "title": title,
            "article_url": article_url,
            "author": author,
            "author_url": author_url,
            "category": category,
            "published": published
        })

    return startups
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_startups():
    driver = webdriver.Chrome()

    driver.get("https://techcrunch.com/category/startups/")
    time.sleep(5)

    titles = driver.find_elements(By.CSS_SELECTOR, "a.loop-card__title-link")
    authors = driver.find_elements(By.CSS_SELECTOR, "a.loop-card__author")

    startups = []

    for title, author in zip(titles, authors):
        startups.append({
            "title": title.text,
            "author": author.text
        })

    driver.quit()

    return startups
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_hackernews():

    driver = webdriver.Chrome()
    driver.get("https://news.ycombinator.com/")
    time.sleep(3)

    news = []

    rows = driver.find_elements(By.CSS_SELECTOR, "tr.athing")

    for row in rows:

        title = row.find_element(By.CSS_SELECTOR, "span.titleline > a").text

        subtext = row.find_element(By.XPATH, "./following-sibling::tr[1]")

        try:
            score = subtext.find_element(By.CSS_SELECTOR, "span.score").text
        except:
            score = "N/A"

        try:
            comments = subtext.find_element(By.XPATH, ".//a[contains(text(),'comment') or text()='discuss']").text
        except:
            comments = "N/A"

        try:
            post_time = subtext.find_element(By.CSS_SELECTOR, "span.age").text
        except:
            post_time = "N/A"

        news.append({
            "title": title,
            "score": score,
            "comments": comments,
            "time": post_time
        })

    driver.quit()
    return news
fetch_hackernews()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_swiggy():

    driver = webdriver.Chrome()
    driver.get("https://www.swiggy.com/")
    time.sleep(8)  # wait for page + popups

    restaurants = []

    # all restaurant cards (best stable container approach)
    cards = driver.find_elements(By.XPATH, "//div[contains(@class,'sc')]")

    for card in cards:

        try:
            name = card.find_element(By.XPATH, ".//div[not(img)]").text
        except:
            name = "N/A"

        try:
            rating = card.find_element(By.XPATH, ".//span[contains(text(),'•') or contains(text(),'.')]").text
            rating = rating.replace("•", "").strip()
        except:
            rating = "N/A"

        try:
            cuisines = card.find_element(By.XPATH, ".//div[contains(text(), ',')]").text
        except:
            cuisines = "N/A"

        try:
            location = card.find_element(By.XPATH, ".//div[contains(text(),'Sector') or contains(text(),'Road') or contains(text(),'Block')]").text
        except:
            location = "N/A"

        restaurants.append({
            "name": name,
            "rating": rating,
            "cuisines": cuisines,
            "location": location
        })

    driver.quit()
    return restaurants
fetch_swiggy()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


def fetch_disasters():

    driver = webdriver.Chrome()
    driver.get("https://www.gdacs.org/")

    time.sleep(8)  # wait for alerts to load

    disasters = []

    # event blocks
    titles = driver.find_elements(By.CSS_SELECTOR, "div.alert_box_title")

    for title in titles:

        try:
            event_type = title.text.strip()
        except:
            event_type = "N/A"

        try:
            parent = title.find_element(By.XPATH, "./following::span[1]")
            location = parent.text
        except:
            location = "N/A"

        try:
            date = title.find_element(By.XPATH, "./following::span[contains(@class,'alert_date')]").text
        except:
            date = "N/A"

        try:
            magnitude = title.find_element(By.XPATH, "./following::span[contains(@class,'magnitude')]").text
        except:
            magnitude = "N/A"

        disasters.append({
            "event": event_type,
            "location": location,
            "date": date,
            "magnitude": magnitude
        })

    driver.quit()
    return disasters
fetch_disasters()
```
